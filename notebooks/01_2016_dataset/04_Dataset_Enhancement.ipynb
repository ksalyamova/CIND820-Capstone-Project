# 2016 Dataset Enhancement

This notebook supplements the existing 2016 ml ready dataset by extracting additional socioeconomic indicators required for the predictive modelling stage.

The objective is to align the 2016 dataset with the standardized feature set used in the 2011 dataset.

```python
from pathlib import Path
import pandas as pd

project_root = Path.cwd().parents[1]

file_path = project_root / "data" / "raw" / "neighbourhood-profiles-2016-140-model.csv"

print(file_path)

df_raw = pd.read_csv(file_path)

print(df_raw.shape)
display(df_raw.head())
```

### 1. Variable Identification

The required socioeconomic indicators are identified within the 2016 Toronto Neighbourhood Profile.

A helper function is used to search the dataset by keyword to verify the availability and location of each research variable before extraction.

```python
def find_indicator(keyword):
    """
    Search the Characteristic column for a keyword.
    """
    results = df_raw[
        df_raw["Characteristic"].str.contains(keyword, case=False, na=False)
    ]

    return results[["_id", "Category", "Topic", "Characteristic"]]
```
