### Methodological Note

During feature extraction, the 2006 Toronto Neighbourhood Profile dataset was examined for neighbourhood-level labour market indicators, including employment rate, participation rate, and unemployment rate.

Although these variables exist in the dataset, they contain values only for the City of Toronto aggregate and are unavailable at the neighbourhood level. Consequently, these variables were excluded from the machine-learning-ready dataset to avoid introducing predictors with predominantly missing values.

The final predictor set therefore consists only of neighbourhood-level variables that are consistently available and suitable for predictive modelling.

### Section 9 – Transform into Machine Learning Dataset

The extracted indicators are currently stored as rows, while neighbourhoods are stored as columns.

To prepare the dataset for predictive modelling, the data structure will be transposed so that:

- each row represents one neighbourhood;
- each column represents one socioeconomic indicator.

This produces the standard machine-learning format where observations are stored as rows and variables are stored as columns.

```python
features_only = selected_df.drop(
    columns=["_id", "Category", "Topic", "Attribute"]
)

features_only.head()
```

```python
ml_df = features_only.T

ml_df.head()
```

```python
ml_df.columns = list(selected_rows.keys())

ml_df.head()
```

```python
ml_df.reset_index(inplace=True)

ml_df.rename(
    columns={"index": "Neighbourhood"},
    inplace=True
)

ml_df.head()
```

## Section 10 – Summary and Next Steps

### Summary

This notebook completed the data preparation and feature extraction process for the 2006 Toronto Neighbourhood Profile dataset.

The raw dataset was explored to identify socioeconomic indicators that could potentially support longitudinal analysis and predictive modelling. Candidate variables were systematically searched, compared with the 2016 dataset, and evaluated based on neighbourhood-level availability and completeness.

The dataset was successfully transformed into a machine-learning format by transposing the neighbourhood profiles so that each row represents a neighbourhood and each column represents a socioeconomic indicator.

The final extracted variables included:

- Population
- Median After-Tax Income
- Employment Rate
- Participation Rate
- Unemployment Rate
- Bachelor's Degree
- Immigrant Population
- Visible Minority Population

### Data Quality Assessment

During the extraction process, an important limitation of the 2006 dataset was identified.

Although labour force and income indicators exist within the dataset, several of these variables contain neighbourhood-level values only for the City of Toronto aggregate while most individual neighbourhood observations are missing.

As a result, variables such as:

- Median After-Tax Income
- Employment Rate
- Participation Rate
- Unemployment Rate

contain substantial missing values and are not suitable for direct use in predictive modelling without introducing significant bias or data loss.

In contrast, variables such as:

- Population
- Bachelor's Degree
- Immigrant Population
- Visible Minority Population

were available for nearly all neighbourhoods and were successfully extracted.

### Conclusion

The 2006 dataset remains valuable for historical comparison and exploratory analysis of neighbourhood socioeconomic conditions. However, its limitations in neighbourhood-level labour market variables make it less suitable as the primary predictor dataset for the planned predictive modelling stage.

Based on these findings, the next phase of the project will evaluate the 2011 Toronto Neighbourhood Profile dataset as a more complete historical baseline for predicting neighbourhood-level socioeconomic outcomes observed in 2016.

This approach is expected to improve data completeness, maintain variable consistency across time periods, and provide a stronger foundation for supervised machine learning models while preserving the 2006 dataset as an important historical reference.
