### 1. Variable Identification

The selected socioeconomic indicators are identified within the 2011 Toronto Neighbourhood Profile.

A helper function is used to quickly search the dataset for variables based on keywords. This allows the selected research variables to be verified before extraction.

```python
def find_indicator(keyword):
    """
    Search the Attribute column for a keyword.
    """

    results = df_raw[
        df_raw["Attribute"].str.contains(keyword, case=False, na=False)    ]

    return results[["Category", "Topic", "Attribute"]]
```

### Feature Extraction

After verifying the selected socioeconomic indicators, the corresponding source rows are extracted from the 2011 Toronto Neighbourhood Profile.

The resulting dataset contains only the variables required for predictive modelling and subsequent preprocessing.

This reduces the original dataset from more than 1,500 indicators to the 10 socioeconomic variables aligned with the research questions.

```python
selected_df = df_raw.loc[list(feature_mapping.values())].copy()

print("=" * 60)
print("FEATURE EXTRACTION SUMMARY")
print("=" * 60)

print(f"\nSelected variables: {selected_df.shape[0]}")
print(f"Columns: {selected_df.shape[1]}")

display(selected_df[["_id", "Category", "Topic", "Attribute"]])
```

### Dataset Transformation

The extracted variables are transformed into a machine-learning-ready dataset.

The original Toronto Neighbourhood Profile stores indicators as rows and neighbourhoods as columns. For predictive modelling, each neighbourhood must represent one observation (row) and each socioeconomic indicator must represent one feature (column).

This section removes the metadata columns, transposes the dataset, and assigns descriptive feature names.

```python
model_data = selected_df.iloc[:, 4:].copy()

print(model_data.shape)

display(model_data.head())
```

```python
model_data.index = feature_mapping.keys()

display(model_data.head())
```

```python
ml_ready = model_data.T

print(ml_ready.shape)

display(ml_ready.head())
```

```python
ml_ready.reset_index(inplace=True)

ml_ready.rename(columns={"index": "Neighbourhood"}, inplace=True)

display(ml_ready.head())
```

### Data Quality Assessment

The transformed dataset is evaluated before export.

The assessment verifies:

- dataset dimensions
- data types
- missing values
- duplicate neighbourhoods

This ensures the modelling dataset is complete and suitable for subsequent exploratory analysis and predictive modelling.

```python
print("=" * 60)
print("DATASET SHAPE")
print("=" * 60)

print(ml_ready.shape)
```

```python
print("=" * 60)
print("DATA TYPES")
print("=" * 60)

display(ml_ready.dtypes)
```

```python
print("=" * 60)
print("MISSING VALUES")
print("=" * 60)

display(ml_ready.isna().sum())
```

```python
print("=" * 60)
print("DUPLICATE NEIGHBOURHOODS")
print("=" * 60)

print(
    ml_ready["Neighbourhood"].duplicated().sum())
```

```python
ml_ready = ml_ready[
    ml_ready["Neighbourhood"] != "City of Toronto"
].reset_index(drop=True)

print("Final dataset shape:")
print(ml_ready.shape)

display(ml_ready.head())
```

### Export ML-Ready Dataset

The cleaned neighbourhood-level dataset is exported as a CSV file for subsequent exploratory data analysis, dimensionality reduction, clustering, and predictive modelling.

This dataset serves as the primary input for the next notebook in the analytical pipeline.

```python
from pathlib import Path

output_dir = project_root / "data" / "processed"
output_dir.mkdir(parents=True, exist_ok=True)

output_file = output_dir / "neighbourhood_profiles_2011_ml_ready.csv"

ml_ready.to_csv(output_file, index=False)

print("=" * 60)
print("EXPORT COMPLETED")
print("=" * 60)
print(f"Dataset exported to:\n{output_file}")
print(f"\nFinal dataset shape: {ml_ready.shape}")
```
