```python
candidate_cols = []

for col in df.columns:
    if "Population, 2016" in col:
        candidate_cols.append(col)

    elif "Median total income of households in 2015 ($)" in col:
        candidate_cols.append(col)

    elif "Median after-tax income of households in 2015 ($)" in col:
        candidate_cols.append(col)

    elif "Government transfers (%)" in col:
        candidate_cols.append(col)

    elif col.endswith("Employment rate"):
        candidate_cols.append(col)

    elif "Employment rate (Males)" in col:
        candidate_cols.append(col)

    elif "Employment rate (Females)" in col:
        candidate_cols.append(col)

    elif col.endswith("No certificate, diploma or degree"):
        candidate_cols.append(col)

    elif col.endswith("Postsecondary certificate, diploma or degree"):
        candidate_cols.append(col)

    elif col.endswith("Bachelor's degree"):
        candidate_cols.append(col)

    elif col.endswith("Master's degree"):
        candidate_cols.append(col)

    elif "Rate of core housing need" in col:
        candidate_cols.append(col)

    elif "Prevalence of low income based on the Low-income measure" in col:
        candidate_cols.append(col)

print("Number found:", len(candidate_cols))

for c in candidate_cols:
    print(c)
```
