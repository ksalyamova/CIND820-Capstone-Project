## Machine Learning Dataset

The evolution of the dataset from the original EDA to revised EDA: The original Toronto Neighbourhood Profiles dataset contained 2,383 indicators in rows and  143 neighbourhoods in columns. After data transposing, we achieved to 143 observations in rows and 2383 neighboorhoods in columns (to float from object)

Data preparation included:
- Removed the metadata fields
- Identified duplicate indicators
- Created unique indicator names (Indicator_v2) 
- Generated revised version of the EDA report (Toronto_EDA_Revised.html)
- Transposed the dataset for machine learning
- Converted values to numeric format
- Created and exported the ML-ready dataset (neighborhood_profiles_ml_ready.csv)
- Documented the transformation process 

Final dataset dimensions:
- Observations (Neighbourhoods): 143
- Features (Indicators): 2,383
