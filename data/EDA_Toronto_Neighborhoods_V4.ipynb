```python
neighbourhood_cols = df.columns[5:]

analysis_df = pd.DataFrame({"Neighbourhood": neighbourhood_cols})

analysis_df.head()
```

```python
def get_values(characteristic):

    row = selected_rows[
        selected_rows["Characteristic_Clean"] == characteristic]

    return row.iloc[0, 5:].values
```

```python
analysis_df["Population_2016"] = get_values("Population, 2016")

analysis_df["Median_Income"] = get_values(
    "Median total income of households in 2015 ($)")

analysis_df["Median_AfterTax_Income"] = get_values(
    "Median after-tax income of households in 2015 ($)")

analysis_df["Government_Transfers_Pct"] = get_values(
    "Government transfers (%)")

analysis_df["Employment_Rate"] = get_values(
    "Employment rate")

analysis_df["Employment_Rate_Male"] = get_values(
    "Employment rate (Males)")

analysis_df["Employment_Rate_Female"] = get_values(
    "Employment rate (Females)")

analysis_df["Core_Housing_Need_Pct"] = get_values(
    "Rate of core housing need")

analysis_df["Low_Income_Pct"] = get_values(
    "Prevalence of low income based on the Low-income measure, after tax (LIM-AT) (%)")
```

# EDA Summary

This exploratory data analysis examines socioeconomic characteristics across Toronto neighbourhoods using selected indicators from the Toronto Neighbourhood Profiles dataset.

The final analysis dataset contains 142 neighbourhoods and the following variables:

- Population (2016)
- Employment Rate
- Employment Rate (Male)
- Employment Rate (Female)
- Low Income Percentage

The dataset was cleaned, transformed into neighbourhood-level observations, converted to numeric formats, and analyzed using YData Profiling to generate an automated EDA report.

The resulting report provides descriptive statistics, distributions, missing-value analysis, and variable relationships to support further socioeconomic analysis.
