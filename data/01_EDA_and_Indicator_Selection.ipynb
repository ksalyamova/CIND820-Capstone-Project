This project investigates socioeconomic vulnerability across Toronto neighbourhoods

### Selected Vulnerability Indicators

After reviewing the available characteristics, twelve indicators were selected to represent major dimensions of socioeconomic vulnerability.
#### Income
- Prevalence of low income (LIM-AT)
#### Education
- No certificate, diploma or degree
- Bachelor's degree
#### Labour Force
- Participation rate
- Employment rate
- Unemployment rate
#### Housing
- Renters
- Rate of core housing need
- Rate of unaffordable housing
- Lone-parent census families
#### Demographic Vulnerability
- Immigrants
- Total visible minority population
These indicators provide broad coverage of economic, educational, housing, labour market, and demographic conditions across Toronto neighbourhoods.

```python
income_vars = df[
    df["Characteristic"].str.contains(
        "income|low income|lim",
        case=False,
        na=False)]

income_vars[["Characteristic"]]
```

### Education Indicators

Educational attainment is associated with employment opportunities,
income potential, and social mobility.

```python
education_vars = df[df["Characteristic"].str.contains(
        "diploma|degree|education",
        case=False,
        na=False)]

education_vars[["Characteristic"]]
```

### Labour Indicators

Labour market participation and unemployment are important measures of
economic opportunity and stability.

```python
labour_vars = df[
    df["Characteristic"].str.contains(
        "employment|unemployment|participation",
        case=False,
        na=False)]

labour_vars[["Characteristic"]]
```

```python
housing_vars = df[
    df["Characteristic"].str.contains(
        "housing|rent|owner",
        case=False,
        na=False)]

housing_vars[["Characteristic"]]
```

### Demographic Indicators

Immigration, visible minority status, and family structure may be associated
with varying levels of socioeconomic vulnerability.

```python
demo_vars = df[
    df["Characteristic"].str.contains(
        "immigrant|minority|visible minority|lone parent",
        case=False,
        na=False)]

demo_vars[["Characteristic"]]
```

```python
selected_indicators = pd.DataFrame({

    "Indicator":[
        "LIM_AT",
        "Immigrants",
        "Visible_Minority",
        "Renters",
        "No_Diploma",
        "Bachelors_Degree",
        "Core_Housing_Need",
        "Unaffordable_Housing",
        "Participation_Rate",
        "Employment_Rate",
        "Unemployment_Rate",
        "Lone_Parent_Families"],

    "Row_ID":[
        1130,
        1152,
        1336,
        1629,
        1703,
        1710,
        1682,
        1688,
        1888,
        1889,
        1890,
        102
    ]})

selected_indicators
```

## Extract Selected Indicators

The selected indicator rows are extracted from the original dataset.

```python
selected_row_ids = selected_indicators["Row_ID"].tolist()

selected_rows = df.loc[selected_row_ids]

selected_rows.head()
```

```python
neighbourhood_data = selected_rows.iloc[:, 6:]

print(neighbourhood_data.shape)
```

## Transpose the Dataset

The dataset is transposed so that:

- Rows represent neighbourhoods
- Columns represent indicators

```python
analysis_df = neighbourhood_data.T

print(analysis_df.shape)
```

## Rename Indicator Columns

Indicator row identifiers are replaced with descriptive variable names.

```python
analysis_df.columns = [
    "LIM_AT",
    "Immigrants",
    "Visible_Minority",
    "Renters",
    "No_Diploma",
    "Bachelors_Degree",
    "Core_Housing_Need",
    "Unaffordable_Housing",
    "Participation_Rate",
    "Employment_Rate",
    "Unemployment_Rate",
    "Lone_Parent_Families"]
```

## Convert Values to Numeric Format

Numeric values are converted from text format into machine-readable numeric variables.

```python
analysis_df = analysis_df.replace(",", "", regex=True)

analysis_df = analysis_df.apply(
    pd.to_numeric,
    errors="coerce")
```

```python
analysis_df.info()
```
