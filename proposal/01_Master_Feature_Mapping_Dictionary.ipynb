# Master Feature Dictionary

## Objective

The Toronto Neighbourhood Profile datasets from 2006, 2011, and 2016 were collected over three census cycles. Although the datasets share a common structure, some indicators differ in wording, availability, or statistical definition.

To ensure methodological consistency and avoid introducing bias into the predictive modelling process, a harmonized feature dictionary was developed before preprocessing and modelling.

Only indicators that were available and conceptually comparable across the three datasets were retained for predictive analysis.

## Target Variable

Neighbourhood Low-Income Prevalence (%)

Reason:
Continuous
Policy relevant
Present in every dataset (with careful harmonization)

Suitable for supervised regression models, including:

- Linear Regression (baseline model)
- Random Forest Regressor (tree-based model)

## Predictor Variables 

1. Population

4. Unemployment Rate

5. Median Income

6. Median After-Tax Income

7. Bachelor Degree

8. Immigrant Population

9. Lone Parent Families

## Methodology Decision

Variables were selected according to five criteria:

1. Availability across the 2006, 2011 and 2016 datasets.
2. Consistent statistical definition.
3. Neighbourhood-level interpretation.
4. Relevance to socioeconomic vulnerability.
5. Support from existing literature.
