## 1. Project Overview

Objective

Develop predictive models to identify socioeconomic factors associated with neighbourhood-level low-income prevalence across Toronto using historical neighbourhood profile data from 2006, 2011, and 2016.

Research Problem

Understanding which socioeconomic indicators best predict future neighbourhood vulnerability can assist policymakers in prioritizing investments and community interventions.

Stakeholders: 

City of Toronto
Urban planners
Housing agencies
Community organizations
Policy analysts

## 2. Research Questions
RQ1

Which socioeconomic indicators are consistently associated with neighbourhood-level low-income prevalence across Toronto?

RQ2

Can socioeconomic characteristics measured in one census period accurately predict neighbourhood-level low-income prevalence in the subsequent census period?

RQ3

Which predictive modelling approach provides the best balance between predictive performance and interpretability?
